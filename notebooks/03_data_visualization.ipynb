```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/cleaned_used_cars.csv")
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,4))
sns.boxplot(x="brand",y='log_transformed_price',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df)
```

```python
df.corr(numeric_only=True)
```
